## 1.	ประเภทของตัวเลขที่ภาษาไพธอนให้การรองรับมีอะไรบ้าง

## 7.จงอธิบายความแตกต่างระหว่างลิสต์และทัพเพิล

```python
# คำตอบ

ลิสต์ (List) เป็นโครงสร้างข้อมูลแบบลำดับที่สามารถ เพิ่ม ลบ หรือแก้ไขค่าได้ (Mutable) ใช้เครื่องหมายวงเล็บเหลี่ยม [] วิธีใช้ เช่น ใช้ [ และ ] เช่น mylist = [1, 2, 3]

ทัพเพิล (Tuple) เป็นโครงสร้างข้อมูลแบบลำดับที่ ไม่สามารถเปลี่ยนแปลงค่าได้ (Immutable) ใช้เครื่องหมายวงเล็บปกติ () เหมาะกับข้อมูลที่ต้องการคงที่ไม่เปลี่ยนแปลง 
วิธีใช้ เช่น ( และ ) เช่น mytuple = (1, 2, 3)
```

## 8.การสลับค่าตัวแปร

```python
"""
a = 10
b = 20

# ผลที่คาดหวังคือ
# a : 20 , b: 10

"""
```

```python
# คำตอบ

a = 10
b = 20

temp = a
a = b
b = temp

print("a: {} , b: {}".format(a, b))
```

```python
# คำตอบ

a = 10
b = 20

# parallel assignment
a, b = b, a

print("a: {} , b: {}".format(a, b))
```
