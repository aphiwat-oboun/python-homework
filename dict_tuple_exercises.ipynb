# แบบฝึกหัด: Dictionary และ Tuple ใน Python

จำนวน 10 ข้อ - สำหรับนักศึกษาที่ต้องการฝึกการใช้งาน Dictionary และ Tuple

## ข้อ 1: สร้าง Dictionary และเข้าถึงค่า

สร้าง dictionary ชื่อ `student` ที่มีข้อมูลดังนี้:
- "name": "สมชาย"
- "age": 20
- "grade": "A"
- "major": "วิทยาศาสตร์คอมพิวเตอร์"

จากนั้นให้พิมพ์ชื่อและเกรดของนักเรียน

````python
# เขียนคำตอบของคุณที่นี่
student = {
    "name": "สมชาย",
    "age": 20,
    "grade": "A",
    "major": "วิทยาศาสตร์คอมพิวเตอร์"
}

print(student.get('name'))
print(student.get('grade'))
````

## ข้อ 2: เพิ่ม แก้ไข และลบข้อมูลใน Dictionary

ให้ใช้ dictionary จากข้อ 1 แล้วทำดังนี้:
- เพิ่มคีย์ใหม่ "city" ที่มีค่า "กรุงเทพ"
- เปลี่ยนอายุเป็น 21
- ลบคีย์ "major" ออก

จากนั้นให้พิมพ์ dictionary ทั้งหมด

````python
# เขียนคำตอบของคุณที่นี่
student = {
    "name": "สมชาย",
    "age": 20,
    "grade": "A",
    "major": "วิทยาศาสตร์คอมพิวเตอร์"
}

student['city'] = 'กรุงเทพ'
student['age'] = 21
student.pop('major')

print(student)
````

## ข้อ 3: การวนลูปผ่าน Dictionary

สร้าง dictionary ชื่อ `prices` ที่เก็บราคาสินค้า:
```python
prices = {
    "apple": 25,
    "banana": 15,
    "orange": 30,
    "mango": 50
}
```

ให้พิมพ์ชื่อสินค้าและราคาในรูปแบบ "แอปเปิล: 25 บาท"

````python
# เขียนคำตอบของคุณที่นี่
prices = {
    "apple": 25,
    "banana": 15,
    "orange": 30,
    "mango": 50
}

for name, price in prices.items():
    print("{0} : {1} บาท".format( name, price ))
````

## ข้อ 4: Tuple - สร้างและเข้าถึงค่า

สร้าง tuple ชื่อ `coordinates` ที่มีค่า (10, 20, 30)

จากนั้นให้พิมพ์:
- ค่าตัวแรก
- ค่าตัวสุดท้าย
- ค่าสามค่า (ใช้ slicing)
- จำนวนสมาชิกทั้งหมด

````python
# เขียนคำตอบของคุณที่นี่
coordinates = (10, 20, 30)

print(coordinates[0])
print(coordinates[-1])
print(coordinates[0:3])
print(coordinates)
````

## ข้อ 5: Tuple ไม่สามารถเปลี่ยนแปลง (Immutable)

สร้าง tuple ชื่อ `colors` ที่มีค่า ("red", "green", "blue")

พยายามเปลี่ยนค่าตัวแรกเป็น "yellow" และบันทึกข้อความ error ที่ได้รับ

จากนั้นให้อธิบายว่าทำไม tuple ถึงไม่สามารถเปลี่ยนแปลงได้

````python
# เขียนคำตอบของคุณที่นี่
colors = ("red", "green", "blue")

colors["red"] = "yellow"

print(colors)

# Tuple เปลี่ยนแปลงไม่ได้ (immutable) เพราะถูกออกแบบให้ค่าภายในคงที่ ไม่ให้แก้ไข เพิ่ม ลบสมาชิก 
# เพื่อความปลอดภัยของข้อมูล ใช้หน่วยความจำมีประสิทธิภาพ และสามารถใช้เป็น key ของ dictionary ได้
````

## ข้อ 6: Dictionary ที่มีค่าเป็น Tuple

สร้าง dictionary ชื่อ `students_scores` ที่เก็บข้อมูลนักเรียน:
```python
students_scores = {
    "นักเรียน1": (85, 90, 88),
    "นักเรียน2": (75, 80, 78),
    "นักเรียน3": (95, 92, 94)
}
```

ให้คำนวณค่าเฉลี่ยคะแนนของนักเรียนแต่ละคน และพิมพ์ผลลัพธ์

````python
# เขียนคำตอบของคุณที่นี่
students_scores = {
    "นักเรียน1": (85, 90, 88),
    "นักเรียน2": (75, 80, 78),
    "นักเรียน3": (95, 92, 94)
}

for name, scores in students_scores.items():
    avg = sum(scores) / len(scores)
    print(f"{name} : ค่าเฉลี่ย = {avg:.2f}")
````

## ข้อ 7: ค้นหาข้อมูลใน Dictionary

สร้าง dictionary ชื่อ `phone_book` ที่เก็บหมายเลขโทรศัพท์:
```python
phone_book = {
    "สมชาย": "081-234-5678",
    "สมหญิง": "089-876-5432",
    "สมศรี": "085-111-2222"
}
```

ให้สร้างฟังก์ชันที่ค้นหาหมายเลขโทรศัพท์ตามชื่อ และถ้าไม่พบให้แสดงข้อความ "ไม่พบข้อมูล"

````python
# เขียนคำตอบของคุณที่นี่
phone_book = {
    "สมชาย": "081-234-5678",
    "สมหญิง": "089-876-5432",
    "สมศรี": "085-111-2222"
}
def search_phone(name):
    if name in phone_book:
        return phone_book[name]
    else:
        return "ไม่พบข้อมูล"
        
print(search_phone("สมชาย"))
print(search_phone("สมหยอง")) 
````

## ข้อ 8: Tuple Unpacking

สร้าง tuple ชื่อ `person_info` ที่มีค่า ("วิชา", 22, "engineer")

ให้ใช้ tuple unpacking เพื่อกำหนดค่าให้กับตัวแปร name, age, และ job

จากนั้นพิมพ์ข้อมูลทั้งหมดในรูปแบบ "วิชา อายุ 22 ปี ประกอบอาชีพ engineer"

````python
# เขียนคำตอบของคุณที่นี่
person_info = ("วิชา", 22, "engineer")

name, age, job = person_info

print(f"ชื่อ: {name} อายุ: {age} ปี ประกอบอาชีพ: {job}")
````

## ข้อ 9: การรวม Dictionary

สร้าง dictionary 2 อัน:
- `person1 = {"name": "สมชาย", "age": 25}`
- `person2 = {"name": "สมหญิง", "age": 24}`

ให้รวม dictionary ทั้งสองเป็น dictionary เดียวชื่อ `all_people` และพิมพ์ผลลัพธ์

หมายเหตุ: ลองใช้วิธีต่างๆ เช่น update(), {**dict1, **dict2} เป็นต้น

````python
# เขียนคำตอบของคุณที่นี่

person1 = {"name": "สมชาย", "age": 25}
person2 = {"name": "สมหญิง", "age": 24}
all_people = person1.copy() 
all_people.update(person2)
print(all_people)

all_people = {**person1,**person2}

print(all_people)
````

## ข้อ 10: โปรแกรมจัดการสินค้า (Challenge)

สร้างโปรแกรมจัดการสินค้าที่มีความสามารถดังนี้:

1. มี dictionary ชื่อ `inventory` ที่เก็บข้อมูลสินค้า: ชื่อสินค้า, จำนวน, และราคา (เก็บในรูป tuple)
2. สร้างฟังก์ชน `add_product()` เพื่อเพิ่มสินค้าใหม่
3. สร้างฟังก์ชั่น `update_quantity()` เพื่ออัปเดตจำนวนสินค้า
4. สร้างฟังก์ชั่น `get_total_value()` เพื่อคำนวณมูลค่ารวมของสินค้า
5. สร้างฟังก์ชั่น `show_inventory()` เพื่อแสดงรายการสินค้าทั้งหมด

ตัวอย่างการใช้งาน:
```
Inventory: {'น้ำตาล': (10, 50), 'แป้งมัน': (5, 100)}
Total value: 1000 บาท
```

````python
# เขียนคำตอบของคุณที่นี่

inventory = {}

def add_product(name, quantity, price):
    inventory[name] = (quantity, price)
    print(f"เพิ่มสินค้า {name} จำนวน {quantity} ราคา {price} บาท")

def update_quantity(name, new_quantity):
    if name in inventory:
        price = inventory[name][1]
        inventory[name] = (new_quantity, price)
        print(f"อัปเดตจำนวน {name} เป็น {new_quantity}")
    else:
        print(f"ไม่พบสินค้า {name}")

def get_total_value():
    total = 0
    for qty, price in inventory.values():
        total += qty * price
    return total

def show_inventory():
    print("\n=== รายการสินค้าในคลัง ===")
    for name, (qty, price) in inventory.items():
        print(f"{name}: จำนวน {qty} ชิ้น, ราคา {price} บาท")
    print("==========================\n")

    add_product("น้ำตาล", 10, 50)
add_product("แป้งมัน", 5, 100)

show_inventory()

total = get_total_value()
print("Total value:", total, "บาท")
````
